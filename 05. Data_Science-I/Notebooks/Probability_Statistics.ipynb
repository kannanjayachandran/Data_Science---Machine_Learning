```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Plotting distributions

- Normal / Gaussian 
- Uniform

```python
# Normal / Gaussian distribution
mean = 0
std_dev = 3
observations = np.random.normal(mean, std_dev, size=1000)
plt.hist(observations, bins=25, density=True, alpha=0.5, color='b')
plt.title('Normal Distribution')
plt.show()

# Percentiles using numpy
# percentiles = np.percentile(observations, np.arange(0, 101)) 
```

```python
# PDF of normal distribution
observations = np.arange(-10, 10, 0.01)
y = stats.norm.pdf(observations, 0, 1)     # scipy normal distribution

plt.plot(observations, y)
plt.title("Gaussian Distribution")
plt.show()
```

```python
# Uniform function

a = 0  # Lower bound
b = 1  # Upper bound

x = np.linspace(a - 0.1, b + 0.1, 1000)  # Create an array of x-values slightly beyond the bounds
pdf = np.full_like(x, 1 / (b - a))  # Uniform PDF

plt.plot(x, pdf, 'g-', lw=2)  # Plot the PDF in red
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)
plt.ylim(0, 1 / (b - a) + 0.1)  # Set the y-axis limits
plt.show()
```

**Q-Q plot**

```python
stats.probplot(observations, dist='norm', plot=plt)
plt.show()
```

**Box Cox Transformation**

$$\color{#F99417}Normal \rightarrow exponential \rightarrow  normal$$

```python
data = 5* np.random.randn(100) + 100

# exponential
exponential = np.exp(data)

# box-cox
res = stats.boxcox(data, 0)

_, axes = plt.subplots(1, 3, figsize=(16, 6))

axes[0].hist(data)
axes[1].hist(exponential)
axes[2].hist(res)

plt.show()
```
