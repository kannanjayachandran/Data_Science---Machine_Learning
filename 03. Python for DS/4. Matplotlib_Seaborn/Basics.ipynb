```python
# importing matplotlib and dependencies

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

- Scatter plot

Scatter plot is a graph in which the values of two variables are plotted along two axes, the pattern of the resulting points revealing any correlation present.

```python
x_data = np.random.random(25) * 100
y_data = np.random.random(25) * 100

plt.scatter(x_data, y_data, c="orange", marker="*", s=120)
```

- We can also tweak the transparency of the points using the `alpha` parameter, which is useful when we have a lot of points to plot and overlap issues. Thus reducing the alpha value will help us to see the overlapping points.

```python
x_data = np.random.random(2500) * 100
y_data = np.random.random(2500) * 100

plt.scatter(x_data, y_data, c="orange", marker="*", s=120, alpha=0.3)
```

- Line chart

A line chart or line graph is a type of chart which displays information as a series of data points called `markers` connected by straight line segments. It is a basic type of chart common in many fields.

```python
years = [x for x in range(1990, 2024)]
stuff_that_change_over_time = [int(x*100) for x in np.random.random(len(years))]
```

```python
plt.plot(years, stuff_that_change_over_time,'--', lw=5)
```

- Bar chart

A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.

```python
ben_10_characters = ["fourArms", "spiderMonkey", "JetRay", "wayBig", "BrainStorm", "DimondHead"]
alien_power = [int(x*100) for x in np.random.random(6)]
```

```python
plt.bar(ben_10_characters, alien_power)
```

- Histogram

A histogram is an accurate representation of the distribution of numerical data. It is an estimate of the probability distribution of a continuous variable (quantitative variable).

```python
mean = 30
std_deviation = 1.5
samples = 1000

salary_per_hr = np.random.normal(mean, std_deviation, samples)
```

```python
plt.hist(salary_per_hr, bins=20)
```

```python
plt.hist(salary_per_hr, bins=20, cumulative=True, color="orange")
```

- 3D Plotting

_They may or may not work in Jupyter Notebook_

```python
ax = plt.axes(projection="3d")

x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter(x,y,z)
ax.set_title("3D Plot")
```

```python
ax = plt.axes(projection="3d")

x = np.arange(0, 50, 0.1)
y = np.arange(0, 50, 0.1)
z = np.cos(x+y)

ax.plot(x,y,z)
ax.set_title("3D Plot")
```
